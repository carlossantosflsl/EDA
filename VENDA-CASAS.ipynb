```python
#Agora, iremos realizar a análise de relação das variáveis por meio de gráficos 
```

```python
sns.histplot(df['SalePrice'], kde=True)
plt.title('Distribuição dos Preços das Casas')
plt.xlabel('Preço')
plt.ylabel('Frequência')
plt.show()
```

```python
#Conseguimos ver a frequência de quantidade de casas pela o seu preço, percebendo que existem uma maior quantidade de casas na 
#faixa de 100000 a 200000.
```

```python
sns.histplot(df['LotArea'], kde=True)
plt.title('Distribuição da Área do Lote')
plt.xlabel('Área do Lote')
plt.ylabel('Frequência')
plt.show()
```

```python
#A maioria das casas a serem vendidas estão entre 50 até mais ou menos 25.000
```

```python
sns.scatterplot(x='LotArea', y='SalePrice', data=df)
plt.title('Preço das Casas vs Área do Lote')
plt.xlabel('Área do Lote')
plt.ylabel('Preço')
plt.show()
```

```python
#Através de um gráfico de dispersão, percebemos existe uma relação entre as áreas dos lotes e os preços das casas 
```

```python
sns.scatterplot(x='YearBuilt', y='SalePrice', data=df)
plt.title('Preço das Casas vs Ano de Construção')
plt.xlabel('Ano de Construção')
plt.ylabel('Preço')
plt.show()
```

```python
#Agora, através do mesmo gráfico, compreendemos que existe casas com preços mais elevados em relação as construções mais recentes. 
#Mas o ano de construção não tem afetado tanto as alterações de preço
```

```python
corr_matrix = df.corr()
top_corr_features = corr_matrix.index[abs(corr_matrix["SalePrice"]) > 0.5]
plt.figure(figsize=(10, 8))
sns.heatmap(df[top_corr_features].corr(), annot=True, cmap='coolwarm')
plt.title('Variáveis mais Correlacionadas com o Preço das Casas')
plt.show()
```

```python
#Agora, neste gráfico de matriz, conseguimos ver a correlação entre as variáveis. Fazendo com que possamos ter uma precisão
#maior em gerar relações entre alterações e modificações de cada variável. 
```

```python
sns.boxplot(x=df['SalePrice'])
plt.title('Boxplot do Preço das Casas')
plt.show()
```

```python
#Agora, através deste gráfico de barra, boxplot, percebemos a mediana dos preços.
```

```python
sns.boxplot(x=df['LotArea'])
plt.title('Boxplot da Área do Lote')
plt.show()
```
